```python
import numpy as np
import scipy.linalg
```

```python
def f(x):
    return np.sin(x/5.) * np.exp(x/10.) + 5 * np.exp(-x/2.) 
```

```python
A = np.array([[1, 1], [1, 15]])
b = np.array([f(1), f(15)])
solve1 = scipy.linalg.solve(A, b)
```

```python
%matplotlib inline
from matplotlib import pylab as plt
```

```python
x = np.arange(0, 15, 0.1)
y = f(x)
```

```python
plt.plot(x, y)
```

```python
y1 = solve1[0] + x*solve1[1]
```

```python
plt.plot(x, y1)
```

```python
A = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b = np.array([f(1), f(8), f(15)])
solve2 = scipy.linalg.solve(A, b)
```

```python
y2 = solve2[0] + x*solve2[1] + x**2 * solve2[2]
```

```python
plt.plot(x, y2)
```

```python
A = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 15**3]])
b = np.array([f(1), f(4), f(10), f(15)])
```

```python
solve3 = scipy.linalg.solve(A, b)
```

```python
y3 = solve3[0] + x*solve3[1] + x**2 * solve3[2] + x**3 * solve3[3]
```

```python
plt.plot(x, y3)
```
